```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoint = np.array([1, 8])
ypoint = np.array([3, 10])
```

```python
plt.plot(xpoint, ypoint, linestyle='-', color='green', marker='o')
plt.show()
```

```python
plt.plot(xpoint, ypoint, linestyle='--')
plt.show()
```

```python
plt.plot(xpoint, ypoint, linestyle='dashdot')
plt.show()
```

```python
plt.plot(xpoint, ypoint, linestyle=':')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(xpoint, ypoint, linestyle=':')
plt.show()
```
