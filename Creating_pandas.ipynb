```python
import pandas as pd

student_dict = {
    'Name': ['Joe', 'Nat', 'Harry'],
    'Age': [20, 21, 19],
    'Marks': [85.10, 77.80, 91.54]
}
```

```python
student_df = pd.DataFrame(student_dict)
print(student_df)
```

```python
student_df = student_df.set_index('Name')
print(student_df)
```

```python
import pandas as pd

data = {
    'Product': ['Laptop', 'Tablet', 'Phone'],
    'Price': [1200, 300, 800],
    'Quantity': [50, 150, 100]
}
df = pd.DataFrame(data)
print(df)
```

```python
df.index = pd.RangeIndex(start=1, stop=4, step=1)

print(df)
```
