```python
data = np.array([85,90,78,92,88,85,95])

data_mean = np.mean(data)
data_median = np.median(data)
data_std = np.std(data)     
data_variance = np.var(data)
data_range = np.max(data) - np.min(data)
```

```python
print(f'mean: {data_mean}')
print(f'median: {data_median}')
print(f'standard deviation: {data_std}')
print(f'variance: {data_variance}') 
print(f'range: {data_range}')
```

```python
temp_data = np.array([22.3, 23.1, 24.5, 25.8, 23.6, 26.7, 27.9, 29.2, 30.5, 24.7, 23.4, 22.1, 25.3, 26.4, 28.7, 29.8, 31.2, 32.4, 30.7, 29.5, 27.8, 26.6])
mean_temp = np.mean(temp_data)
median_temp = np.median(temp_data)
std_dev_temp = np.std(temp_data)

print(f'mean: {mean_temp}')
print(f'median: {median_temp}')
print(f'standard deviation: {std_dev_temp}')
```

```python
temp_in_f = (temp_data * 9/5) + 32
print(temp_in_f)
```

```python
hot_days = temp_data[temp_data > 30]
print('Hot Temp Days:', hot_days)
print('Num of Hot Days:', len(hot_days))
```

```python
base_temp = 20
cooling_degrees_total = np.sum(np.maximum(0, base_temp - temp_data))
print('Cooling Degrees:', cooling_degrees_total)
```
